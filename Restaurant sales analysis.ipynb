## Data Cleaning

```python
# Drop columns that are not required for analysis
df.drop(['Instructions', 'Order Ready Marked', 'Delivery', 'Customer complaint tag', 'Restaurant penalty (Rejection)'], axis=1, inplace=True)
```

```python
df.drop(['Cancellation / Rejection reason'], axis=1, inplace=True)
```

```python
df.drop(['Discount construct','Restaurant compensation (Cancellation)','Review','City','Restaurant discount (Promo)','Gold discount','Brand pack discount'], axis=1, inplace=True)
```

```python
df.columns
```

```python
df.rename(columns={'Restaurant discount (Flat offs, Freebies & others)' : 'Restaurant discount'})
```

## FINDING THE NULL VALUE

```python
df.isnull().sum()
```

LET US ANALYSE RATING COLUMN

```python
df['Rating'].sample(20)
```

```python
df['Rating'] = df['Rating'].fillna(df.groupby('Restaurant name')['Rating'].transform('median'))

```

ANALYSE KPT NULL VALUES

```python
df['KPT duration (minutes)'].sample(10)
```

## REPLACE KPT NAN VALUES BY MEDIAN TIME OF THAT RESTAURANT

```python
df['KPT duration (minutes)'] = df['KPT duration (minutes)'].fillna(df.groupby('Restaurant name')['KPT duration (minutes)'].transform('median'))
```

```python
df['KPT duration (minutes)'].sample(10)
```

## Rider wait time nan values are being replaced by median time

```python
df['Rider wait time (minutes)']=df['Rider wait time (minutes)'].fillna(df.groupby('Restaurant name')['Rider wait time (minutes)'].transform('median'))
```

```python
df['Rider wait time (minutes)'].sample(15)
```

## Exploratory Data Analysis (EDA)

## Distribution of Ratings:

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Rating'], bins=20, kde=True , color='red')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

TOP RESTAURANTS

```python
# Get top 6 most common restaurant names
top6 = df['Restaurant name'].value_counts().nlargest(6).index

# Filter dataset for only these top 6
df_top = df[df['Restaurant name'].isin(top6)]

plt.figure(figsize=(12, 8))
sns.countplot(
    y='Restaurant name',
    data=df_top,
    order=top6,
    palette='tab10')

plt.title('Top 6 Restaurant Names', fontsize=14, fontweight='bold')
plt.xlabel('Count')
plt.ylabel('Restaurant Name')

plt.yticks(fontsize=10)
plt.show()
```
